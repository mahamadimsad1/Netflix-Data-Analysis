```python
# we would cut the Vote_Average values and make 4 categories popular average below_average not popular to describe it more using categories_col function() function provide above
```

```python
import pandas as pd

def categorize_col(df, col, labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
```

```python
labels = ['not popular', 'below_average', 'average', 'popular']
categorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```
